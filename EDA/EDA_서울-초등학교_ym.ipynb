```python
skid = pd.read_csv('./skid.csv')
```

```python
pre = pd.read_excel("./school2019.xlsx", encoding='ansi')
pre.head(3)
```

```python
len(pre)
```

```python
# 필요한 열만 남기자
pre2 = pre.filter(['학교명', '학교급구분', '소재지도로명주소', '시도교육청명' ,'위도', '경도'])
pre2
```

```python
#초등학교만 추출
pre3 = pre2[pre2['학교급구분']=='초등학교']
#서울초등학교
pre4 = pre3[pre3['시도교육청명']=='서울특별시교육청']
```

```python
len(pre3)
```

```python
len(pre4)
```

```python
# 사고다발 & 초등학교 겹치기
# zoom_start: 배율 1~22
map_osm = folium.Map(location=[37.566651, 126.978428], zoom_start=11.3)
map_osm

for i in skid.index:
    lat = skid.loc[i, '위도']
    lon = skid.loc[i, '경도']
    name = skid.loc[i, '사고지역위치명']
    ser = skid.loc[i, '심각도']
    marker = folium.CircleMarker([lat,lon], radius=int(ser), color='#FFb90F', fill_color='#FFb90F',alpha=1, popup=name)
    marker.add_to(map_osm)

    
for i in pre4.index:
    lat = pre4.loc[i, '위도']
    lon = pre4.loc[i, '경도']
    name = pre4.loc[i, '학교명']
    marker = folium.CircleMarker([lat,lon], radius=1, color='#3186cc', fill_color='#3186cc',alpha=1, popup=name)
    marker.add_to(map_osm)
    
map_osm #징그러워... 
```
