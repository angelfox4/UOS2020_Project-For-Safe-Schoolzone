```python
data1 = pd.read_csv("C:/Users/82104/Downloads/스쿨존검색어.csv")
```

```python
d1 = data1[['sup_word', 'sub_word1']]
d2 = data1[['sub_word1', 'sub_word2']]
d3 = data1[['sub_word2', 'sub_word3']]
```

```python
d1.columns = ['source','target']
d2.columns = ['source','target']
d3.columns = ['source','target']
```

```python
df = pd.concat([d1,d2])
df = pd.concat([df,d3])
df
```

### 1. node df 생성

```python
a = df[['source']]
b = df[['target']]

a.columns = ['node']
b.columns = ['node']

node = pd.concat([a,b])
node = node.reset_index(drop=True, inplace=False)
```

```python
# 중복 제거
node = node.drop_duplicates()
```

```python
node = node.reset_index(drop=True, inplace=False)
```

```python
node['idx'] = node.index
```

```python
node
```

```python
node.to_csv('./node_sczone.csv')
```

```python
node_dict = node.set_index('node').T.to_dict('list')
```

```python
node_dict
```

### 2. link df

```python
df['source_idx'] = df['source'].replace(node_dict)
```

```python
df['target_idx'] = df['target'].replace(node_dict)
```

```python
link = df
```

```python
link.to_csv('./link_sczone.csv')
```
